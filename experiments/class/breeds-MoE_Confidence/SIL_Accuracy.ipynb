```python
import numpy as np
import json
import os

def compute_accuracies(matrix):
    matrix = np.array(matrix)
    n = len(matrix)
    assert matrix.shape == (n, n), "Matrix must be NxN"
    seen_means = []
    unseen_means = []
    row_means = []
    print("Seen Accuracy")
    for i in range(n):
        seen_values = matrix[i][: i + 1].tolist()
        mean_val = np.mean(seen_values)
        seen_means.append(mean_val)
        print(f"  Step {i}: {mean_val:.4f}")
    overall_seen = np.mean(seen_means)
    print("\nUnseen Accuracy")
    for i in range(n):
        unseen_values = matrix[i][i + 1 :].tolist()
        if len(unseen_values) != 0:
            mean_val = np.mean(unseen_values)
            unseen_means.append(mean_val)
            print(f"  Step {i}: {mean_val:.4f}")
    overall_unseen = np.mean(unseen_means)
    for i in range(n):
        row_mean = np.mean(matrix[i])
        row_means.append(row_mean)
    overall_all = np.mean(row_means)
    print(f"\nAll Accuracy: {overall_all:.4f}\n")


def load_matrix_from_json(filepath):
    print(os.path.basename(filepath))
    rows = []
    with open(filepath, "r") as f:
        for line in f:
            line = line.strip()
            if not line:
                continue
            entry = json.loads(line)
            if "acc_per_task" in entry:
                rows.append(entry["acc_per_task"])
    n = len(rows)
    matrix = [row[:n] for row in rows]
    compute_accuracies(matrix)


if __name__ == "__main__":
    matrix = load_matrix_from_json("metrics_13-5.json")
    matrix = load_matrix_from_json("metrics_13-10.json")
    matrix = load_matrix_from_json("metrics_13-13.json")
    matrix = load_matrix_from_json("metrics_30-4.json")
    matrix = load_matrix_from_json("metrics_30-8.json")
    matrix = load_matrix_from_json("metrics_30-15.json")
```
